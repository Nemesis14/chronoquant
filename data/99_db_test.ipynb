```python
# =============================================================================
# simple db test: select the last 5 rows from sqlite table
# =============================================================================
import sqlite3
import pandas as pd

# =============================================================================
# database path and table name
# =============================================================================
db_path = "bchusdt_data.db"
table   = "bchusdt_1m"

# =============================================================================
# connect and select last 5 rows (ordered by open_time_ms desc)
# =============================================================================
conn   = sqlite3.connect(db_path)
query  = f"""
    select *
    from {table}
    order by open_time_ms desc, open_time desc
    limit 5
"""
df     = pd.read_sql_query(query, conn)
conn.close()

# =============================================================================
# display result
# =============================================================================
print("last 5 rows (most recent first):")
display(df)
```
